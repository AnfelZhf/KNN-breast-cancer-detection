# Data vizualization 

```python
X = df.drop(columns=['id', 'diagnosis'])

# Sélection de la variable cible
y = df['diagnosis']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Tracer le nombre d'instances dans chaque classe
plt.figure(figsize=(2, 5))
sns.countplot(x=y, palette='Set1')
plt.title('Nombre d\'instances dans chaque classe du cancer du sein')
plt.xlabel('Classe')
plt.ylabel('Nombre d\'instances')
plt.show()
```

# Data cleaning

```python
df['diagnosis'].replace({'M': 'Malignant', 'B': 'Benign'}, inplace=True)
```

# Data normalization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# Fit and transform the feature matrix (X)
X_normalized = scaler.fit_transform(X)

# Display the mean and standard deviation of the normalized features
print("Mean of normalized features:")
print(X_normalized.mean(axis=0))
print("\nStandard deviation of normalized features:")
print(X_normalized.std(axis=0))
```

# Data split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=44, shuffle =True)
```

# Initialization of the KNN Model

```python
import matplotlib.pyplot as plt

# Définir une liste de valeurs de k
neighbors = range(1,10)  # Par exemple, on teste de 1 à 20 voisins

# Initialiser une liste pour stocker les scores d'exactitude
accuracy_scores = []

# Boucle à travers différentes valeurs de k
for k in neighbors:
    # Créer et entraîner le modèle KNN
    KNNClassifierModel = KNeighborsClassifier(n_neighbors=k, weights='uniform')
    KNNClassifierModel.fit(X_train, y_train)
    
    # Calculer et stocker le score d'exactitude sur l'ensemble de test
    accuracy = KNNClassifierModel.score(X_test, y_test)
    accuracy_scores.append(accuracy)

# Tracer le graphique
plt.plot(neighbors, accuracy_scores, marker='o')
plt.title('Exactitude du modèle KNN en fonction du nombre de voisins')
plt.xlabel('Nombre de voisins (k)')
plt.ylabel('Exactitude')
plt.xticks(neighbors)  # Assurez-vous que les étiquettes sur l'axe x représentent les valeurs de k
plt.grid(True)
plt.show()
```

```python
KNNClassifierModel = KNeighborsClassifier(n_neighbors= 4,weights ='uniform',#can be distance
                                          algorithm='auto') # it can be ball_tree, kd_tree,brute
```

# Training

```python
KNNClassifierModel.fit(X_train, y_train)
```

# Prediction

```python
y_pred = KNNClassifierModel.predict(X_test)
y_pred_prob = KNNClassifierModel.predict_proba(X_test)
print('Predicted Value for KNNClassifierModel is : ' , y_pred[:10])
print('Prediction Probabilities Value for KNNClassifierModel is : ' , y_pred_prob[:10])
```

# Evaluation

```python
import matplotlib.pyplot as plt

# Définir une liste de valeurs de k
neighbors = range(1,10)  # Par exemple, on teste de 1 à 20 voisins

# Initialiser une liste pour stocker les scores d'exactitude
accuracy_scores = []

# Boucle à travers différentes valeurs de k
for k in neighbors:
    # Créer et entraîner le modèle KNN
    KNNClassifierModel = KNeighborsClassifier(n_neighbors=k, weights='uniform')
    KNNClassifierModel.fit(X_train, y_train)
    
    # Calculer et stocker le score d'exactitude sur l'ensemble de test
    accuracy = KNNClassifierModel.score(X_test, y_test)
    accuracy_scores.append(accuracy)

# Tracer le graphique
plt.plot(neighbors, accuracy_scores, marker='o')
plt.title('Exactitude du modèle KNN en fonction du nombre de voisins')
plt.xlabel('Nombre de voisins (k)')
plt.ylabel('Exactitude')
plt.xticks(neighbors)  # Assurez-vous que les étiquettes sur l'axe x représentent les valeurs de k
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import classification_report
print("Classification Report:")
print(classification_report(y_test, y_pred))
```
